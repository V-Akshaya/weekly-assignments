## 1. Read the dataset to the python environment.

```python
import pandas as pd
cdata=pd.read_csv('cars_data.csv')
cdata
```

## 2. Check for the null values present in the dataset.


```python
cdata.isnull().sum()
```

There is no null value is present in the dataset.

## 3. Plot a bar graph of male vs female buyers participated in the sales.

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(5,6))
gender = cdata['Buyer Gender'].value_counts().to_frame().reset_index()
plt.bar(gender['index'],gender['Buyer Gender'],color = ['maroon','#B500FF'])
plt.title("MALE VS FEMALE")
plt.xlabel('Gender')
plt.ylabel('Count')
```

#### INSIGHT

1. In this graph we can clearly see that the female buyers is little higher than the male buyers.
2. Driven by education,work,higher spending power and changing lifestyles , an increasing number of 
   women are taking to the wheel.
3. Car buying is considered to be a male-dominated arena, but this graph clearly shows the changing
   trend of buyers in the world.


## 4. Find the top 5 cars based on their sales price.


Finding the top 5 cars by sorting the data in descending order based on sales price.

```python
cdata.sort_values('Sale Price',ascending=False).head(5)
```

Finding the top 5 cars by using nlargest() based on the sale price.

```python
cdata.nlargest(5, "Sale Price").reset_index()
```

```python
#Another way of finding Top 5 car price 
cdata['Sale Price'].nlargest(n=5)
```

#### The top 5 cars based on the Sale price are:
- Volkswagen
- Audi
- Lexus A6
- Mazda
- Lexus ES

## 5. Find the least 5 cars based on their Resell price

Finding the least 5 cars by sorting the data based Resell price

```python
cdata.sort_values('Resell Price').head(5)
```

Finding the least 5 cars by using nsmallest() based on the Resell price.

```python
cdata.nsmallest(5, "Resell Price").reset_index()
```
