## *5. Perform meaningful visualizations using the dataset. Bring at least 3 visualization.*

# BOXPLOT

```python
#Boxplot for iris dataset taking petal length as feature
import seaborn as sns
sns.boxplot(x='Classification',y='PL',data=iris_data)
```

# VIOLIN PLOT

```python
#Violin plot for iris dataset taking petal length as feature
```

```python
sns.violinplot(x='Classification',y='PL',data=iris_data)
```

# Pair Plot

```python
#pairplot for iris dataset shows correlation between each class.
```

```python
sns.pairplot(iris_data)
```

# Scatter Plot

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))
plt.subplot(2,1,1)
plt.scatter(iris_data['PL'],iris_data['PW'],s=10,c="green",marker='^')
plt.title("Scatter plot of petal length vs petal width",fontsize=16)
plt.xlabel("Petal length",fontsize=10)
plt.ylabel("Petal width",fontsize=10)

plt.subplot(2,1,2)
plt.scatter(iris_data['SL'],iris_data['SW'],s=10,c="red",marker='^')
plt.title(" Scatter plot of sepal length vs sepal width",fontsize=16)
plt.xlabel("Sepal length",fontsize=10)
plt.ylabel("Sepal width",fontsize=10)

plt.tight_layout()
```

# Histogram 

```python
#histogram of petal length
```

```python
plt.figure(figsize=(8,8))
plt.hist(iris_data['PL'],color='purple',rwidth=0.6,density=True)
plt.title("Iris Petal Length",fontsize=16)
plt.xlabel("Petal length")
plt.ylabel("Frequency")
plt.xticks()
plt.yticks()
```

```python
#Histogram of petal length vs sepal length
```

```python
%matplotlib inline
plt.figure(figsize=(8,6))
plt.hist(iris_data["PL"],label="Petal length")
plt.hist(iris_data["SL"],label="Sepal length")
plt.xlabel("Data", size=14)
plt.title("Petal length VS Sepal length")
plt.legend(loc='upper right')
plt.savefig("overlapping_histograms_with_matplotlib_Python.png")
```
